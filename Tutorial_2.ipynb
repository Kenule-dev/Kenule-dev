```python
import numpy as np
import pandas as pd
from numpy import random

x1=np.array(([1, 2, 3], [5,5,7], [12, 23, 9])); print(x1)
print(np.linalg.det(x1))

y1=np.random.randint(10, 20, size=(3, 3)); y2=np.random.randint(15, 40, size=(3, 3))
y_total= np.add(y1, y2); print(np.linalg.det(y_total))
print(np.linalg.inv(y_total))
```
